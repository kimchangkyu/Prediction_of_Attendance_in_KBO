## 팀 순위

```python
rank_df = pd.read_csv("datas/kbo_rank.csv")
```

```python
rank_df.tail()
```

```python
#팀별 순위 데이터 프레임 생성

hw = pd.DataFrame(rank_df[["date", "hw"]])
hw["team"] = "한화"
hw.columns = ["date", "hw_rank", "team"]

kia = pd.DataFrame(rank_df[["date", "kia"]])
kia["team"] = "KIA"
kia.columns = ["date", "kia_rank", "team"]

kt = pd.DataFrame(rank_df[["date", "kt"]])
kt["team"] = "KT"
kt.columns = ["date", "kt_rank", "team"]

lg = pd.DataFrame(rank_df[["date", "lg"]])
lg["team"] = "LG"
lg.columns = ["date", "lg_rank", "team"]

ld = pd.DataFrame(rank_df[["date", "ld"]])
ld["team"] = "롯데"
ld.columns = ["date", "ld_rank", "team"]

nc = pd.DataFrame(rank_df[["date", "nc"]])
nc["team"] = "NC"
nc.columns = ["date", "nc_rank", "team"]

dusan = pd.DataFrame(rank_df[["date", "ob"]])
dusan["team"] = "두산"
dusan.columns = ["date", "dusan_rank", "team"]

sk = pd.DataFrame(rank_df[["date", "sk"]])
sk["team"] = "SK"
sk.columns = ["date", "sk_rank", "team"]

ss = pd.DataFrame(rank_df[["date", "ss"]])
ss["team"] = "삼성"
ss.columns = ["date", "ss_rank", "team"]

kw = pd.DataFrame(rank_df[["date", "kw"]])
kw["team"] = "키움"
kw.columns = ["date", "kw_rank", "team"]
```

#### 경기결과에 홈팀기준 순위추가

```python
# lg
kbo = pd.merge(kbo, lg, how="left", left_on=[
    "date", "home"], right_on=["date", "team"])
kbo = kbo.drop(["team"], axis=1)

# dusan
kbo = pd.merge(kbo, dusan, how="left", left_on=[
    "date", "home"], right_on=["date", "team"])
kbo = kbo.drop(["team"], axis=1)

kbo = kbo.fillna(value=0)
kbo["home_rank"] = kbo["lg_rank"] + kbo["dusan_rank"]
kbo = kbo.drop(["lg_rank", "dusan_rank"], axis=1)
```

```python
kbo
```

#### 경기결과에 원정팀기준 순위추가

```python
df = pd.merge(kbo, hw, how="left", left_on=[
    "date", "away"], right_on=["date", "team"])
df = df.drop(["team"], axis=1)

df = pd.merge(df, kia, how="left", left_on=[
    "date", "away"], right_on=["date", "team"])
df = df.drop(["team"], axis=1)

df = pd.merge(df, kt, how="left", left_on=[
    "date", "away"], right_on=["date", "team"])
df = df.drop(["team"], axis=1)

df = pd.merge(df, lg, how="left", left_on=[
    "date", "away"], right_on=["date", "team"])
df = df.drop(["team"], axis=1)

df = pd.merge(df, ld, how="left", left_on=[
    "date", "away"], right_on=["date", "team"])
df = df.drop(["team"], axis=1)

df = pd.merge(df, nc, how="left", left_on=[
    "date", "away"], right_on=["date", "team"])
df = df.drop(["team"], axis=1)

df = pd.merge(df, dusan, how="left", left_on=[
    "date", "away"], right_on=["date", "team"])
df = df.drop(["team"], axis=1)

df = pd.merge(df, sk, how="left", left_on=[
    "date", "away"], right_on=["date", "team"])
df = df.drop(["team"], axis=1)

df = pd.merge(df, ss, how="left", left_on=[
    "date", "away"], right_on=["date", "team"])
df = df.drop(["team"], axis=1)

df = pd.merge(df, kw, how="left", left_on=[
    "date", "away"], right_on=["date", "team"])
df = df.drop(["team"], axis=1)
```

```python
df = df.fillna(value=0)

df["away_rank"] = df['hw_rank'] + df['kia_rank'] + df['kt_rank'] + df['lg_rank'] + \
    df['ld_rank'] + df['nc_rank'] + df['dusan_rank'] + \
    df['sk_rank'] + df['ss_rank'] + df['kw_rank']

df = df.drop(['hw_rank',
              'kia_rank', 'kt_rank', 'lg_rank', 'ld_rank', 'nc_rank', 'dusan_rank',
              'sk_rank', 'ss_rank', 'kw_rank'], axis=1)

df['home-away'] = abs(df['home_rank'] - df['away_rank'])

df = df.drop(["away_rank"], axis=1).sort_values("date")
```

```python
df
```

#### 홈팀 결과 및 순위 shift
- 앞 경기 결과를 예측에 사용하기 위해서

```python
dusan = df[df["home"] == "두산"]
dusan = dusan.sort_values("date")
dusan["win_shift"] = dusan["homewin"].shift(1)
dusan["rank_shift"] = dusan["home_rank"].shift(1)

lg = df[df["home"] == "LG"]
lg = lg.sort_values("date")
lg["win_shift"] = lg["homewin"].shift(1)
lg["rank_shift"] = lg["home_rank"].shift(1)

df = pd.concat([dusan, lg]).sort_values("date")
```

```python
df
```

```python
df["win_shift"] = df["win_shift"].fillna(value="same")
df['rank_shift'] = df['rank_shift'].fillna(value=0)
df = df.drop(["home_rank", "park", "etcs", "away", "times"], axis=1)
```

### 누적승수

#### 두산

```python
dusan = df[df["home"] == "두산"]
```

```python
dusan["win_shift"] = dusan["win_shift"].apply(lambda data: 1 if data == "win" else 0)
```

```python
li_2017 = []
li_2018 = []
li_2019 = []

count_2017 = 0
count_2018 = 0
count_2019 = 0

for win, date in zip(dusan["win_shift"].values, dusan["date"].values):
    if date[:4] == "2017":
        count_2017 += win
        li_2017.append(count_2017)
    elif date[:4] == "2018":
        count_2018 += win
        li_2018.append(count_2018)
    elif date[:4] == "2019":
        count_2019 += win
        li_2019.append(count_2019)
```

```python
win = li_2017 + li_2018 + li_2019
dusan["cumsum_win"] = win
```

#### LG

```python
lg = df[df["home"] == "LG"]
```

```python
lg["win_shift"] = lg["win_shift"].apply(lambda data: 1 if data == "win" else 0)
```

```python
li_2017 = []
li_2018 = []
li_2019 = []

count_2017 = 0
count_2018 = 0
count_2019 = 0

for win, date in zip(lg["win_shift"].values, lg["date"].values):
    if date[:4] == "2017":
        count_2017 += win
        li_2017.append(count_2017)
    elif date[:4] == "2018":
        count_2018 += win
        li_2018.append(count_2018)
    elif date[:4] == "2019":
        count_2019 += win
        li_2019.append(count_2019)
```

```python
win = li_2017 + li_2018 + li_2019
lg["cumsum_win"] = win
```

### 연승지수

```python
def win_continue(team):
    df2 = df[df["home"] == team]
    df2["win"] = df2["homewin"].apply(lambda x: 1 if x == "win" else 0)
    df2["lose"] = df2["homewin"].apply(lambda x: 1 if x == "lose" else 0)
    m = df2["win"].astype(bool)
    df2['win_streak'] = (
        m.groupby([m, (~m).cumsum().where(m)]).cumcount().add(1).mul(m))
    m = df2["lose"].astype(bool)
    df2['lose_streak'] = (
        m.groupby([m, (~m).cumsum().where(m)]).cumcount().add(1).mul(m))
    df2["streak"] = df2["win_streak"] + df2["lose_streak"]*-1
    df2 = df2.drop(["win_streak", "lose_streak", "win", "lose"], axis=1)
    return df2
```

```python
dusan = win_continue("두산")
lg = win_continue("LG")

concat_df = pd.concat([dusan, lg])
df = concat_df.sort_values(by="date").reset_index(drop=True)
df = df.drop(["homewin"], axis=1)
```
