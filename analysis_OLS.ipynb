## 1. 배경

&nbsp;&nbsp;&nbsp;&nbsp;한국프로야구 관중수 즉 티켓판매 비용은 프로구단의 가장 큰 수입원 가운데 하나이다. 이에 따라, 직,간접적인 지표를 바탕으로 프로야구 관중수를 예측할 수 있다면, 그에 맞는 관중 유치 전략을 세우는데 도움을 줄 수 있다.  

## 2. 관련 연구 및 배경지식
### 2.1 관람에 영향을 주는 요인
&nbsp;&nbsp;&nbsp;&nbsp; 특정 행동에 대한 보상이 따를 경우 행동의도가 높아 질 수 있다는 강화이론에 따르면 팬의 관심과 팀의 성적을 바탕으로 야구 관중 수 또한 증가할 수 있다는 것을 알 수 있다. 아래 그림을 통해 위 내용을 확인 할 수 있다. 다만, 관심이 높고 보상이 만족스러워도 날씨가 악화되거나, 시간적 여유가 없는 경우 야구장을 찾기 힘들다.

&nbsp;&nbsp;&nbsp;&nbsp; 그 결과 지금까지 관중 수에 관심을 가진 연구들 모두 팀의 성적과 환경 변수 등을 바탕으로 예측 모델을 제안해왔다. 다만, 야구 팬의 관심을 반영한 데이터는 KBO 홈페이지에서 수치적으로 명확히 제공하는 팀의 성적과 달리 수집이 어렵기 때문에 반영되지 못한 경우가 많다. 물론, 박진욱, & 박상현. (2017)의 연구에서 포탈사이트 네이버에서 제공하는 경기별 댓글 데이터를 수집하여 예측모델에 반영하였으나, 해당 댓글은 게시수도 많지 않을 뿐만 아니라, 광고성 글 등이 많이 포함되어 정확한 관심을 반영한 데이터로 볼 수 없다.  

&nbsp;&nbsp;&nbsp;&nbsp; 그래서 야구 커뮤니티 사이트 엠팍을 선택했다. 다행히, 엠팍에서는 아래 그림과 같이 관심있는 팀을 지정하고 그 팀에 대한 자신의 의견을 반영하는 게시판 서비스를 제공하고 있다. 이를 활용할 경우 각 팀에 대한 관심도를 반영할 수 있고 광고성 글을 걸러 낼 수 있을 뿐만 아니라 각 팀에 대한 게시물이 하루 평균 약 850건이 게시되어 그 양 또한 풍부하다고 볼 수 있다.

### 2.2 분석방법 
&nbsp;&nbsp;&nbsp;&nbsp;특정변수 값을 입력 받아 출력변수 값을 계산해내는 회귀 분석을 활용하여 본 연구의 목적을 해결하고자한다. 비교 대상이되는 사전 연구에서 MAPE값을 활용하여 모델의 설명력을 제안함에 따라, 본 연구에서도 MAPE값을 활용하여 모델을 설명하고자 한다.

$$ M(MAPE) = {100 \over n} {\sum_{k=1}^N \left| {\widehat{y}-y \over \widehat{y}}\right|}$$

&nbsp;&nbsp;&nbsp;&nbsp; 아래 그림은 사전연구에서 제안한 MAPE 설명력 예시이다.

````python
model1 = sm.OLS.from_formula("attendance ~ C(month) + C(weekday) + C(weekend) + C(rainy) + scale(mean_c) + scale(O3) + scale(PM10) + scale(PM25) + C(home) + scale(home_away) + C(win_shift) + scale(rank_shift) + scale(streak) + scale(cumsum_win) + scale(m_count) + scale(search)", data=df)
result1 = model1.fit()
print(result1.summary())
````

- MAPE

````python
pred = result1.predict(df)
ls = []
for num in range(len(pred)):
    ls.append(np.abs((df.attendance.reset_index(drop=True)[
              num] - pred.reset_index(drop=True)[num]) / df.attendance.reset_index(drop=True)[num]))
result_t_train = np.mean(ls)*100
print("MAPE: ", round(result_t_train, 2))
````

- k 폴드 교차검증

````python
from sklearn.model_selection import KFold
scores = np.zeros(10)
cv = KFold(5, shuffle=True, random_state=22)
total_ls = []
for i, (idx_train, idx_test) in enumerate(cv.split(df)):
    df_train = df.iloc[idx_train]
    df_test = df.iloc[idx_test]
    model1 = sm.OLS.from_formula("attendance ~ C(month) + C(weekday) + C(weekend) + C(rainy) + scale(mean_c) + scale(O3) + scale(PM10) + scale(PM25) + C(home) + scale(home_away) + C(win_shift) + scale(rank_shift) + scale(streak) + scale(cumsum_win) + scale(m_count) + scale(search)", data=df)
    result1 = model1.fit()
    pred = result1.predict(df_train)
    ls = []
    for num in range(len(pred)):
        ls.append(np.abs(1 - pred.reset_index(drop=True)
                         [num] / df_train.attendance.reset_index(drop=True)[num]))
    result_t_train = np.mean(ls)
    pred = result1.predict(df_test)
    ls = []
    for num in range(len(pred)):
        ls.append(np.abs(1 - pred.reset_index(drop=True)
                         [num] / df_test.attendance.reset_index(drop=True)[num]))
    result_t_test = np.mean(ls)
    print("학습 error = {:.4f}, 검증 error = {:.4f}".format(
        result_t_train, result_t_test))
````

- 카테고리형과 실수형 변수간 특성

````python
def corr_group(df, var_1, var_2, group_list):
    # correlaiton fuction with 2 variables
    def corr_func(g): return g[var_1].corr(g[var_2])
    # GroupBy operator
    grouped = df.groupby(group_list)
    # calculate correlation coefficient by Group
    corr_coef_df = pd.DataFrame(
        grouped.apply(corr_func), columns=['corr_coef'])
    # add var_1, var_2 column names
    corr_coef_df['var1'] = var_1
    corr_coef_df['var2'] = var_2
    return corr_coef_df
````

````python
# 월:누적승수, 월:게시글 갯수, 월:검색량, 요일:누적승수, 요일:게시글 갯수, 요일:검색량
win = corr_group(df, "attendance", "cumsum_win", "month")
win.reset_index(inplace=True)
win["month"] = win["month"].astype(int)
win = win.sort_values("month")

count = corr_group(df, "attendance", "m_count", "month")
count.reset_index(inplace=True)
count["month"] = count["month"].astype(int)
count = count.sort_values("month")

search = corr_group(df, "attendance", "search", "month")
search.reset_index(inplace=True)
search["month"] = search["month"].astype(int)
search = search.sort_values("month")

win_k = corr_group(df, "attendance", "cumsum_win", "weekend")

count_k = corr_group(df, "attendance", "m_count", "weekend")

search_k = corr_group(df, "attendance", "search", "weekend")
````

````python
plt.figure(figsize=(15, 8))
plt.subplot(231)
plt.plot(win["month"], win["corr_coef"])
plt.title("월별 관중수와 연승지수 사이 상관관계")
plt.ylabel("상관계수")

plt.subplot(232)
plt.plot(count["month"], count["corr_coef"])
plt.title("월별 관중수와 게시글 갯수 사이 상관관계")

plt.subplot(233)
plt.plot(search["month"], search["corr_coef"])
plt.title("월별 관중수와 팀 검색량 사이 상관관계")

plt.subplot(234)
plt.plot(win_k.index, win_k["corr_coef"])
plt.title("휴일별 관중수와 연승지수 사이 상관관계")
plt.ylabel("상관계수")

plt.subplot(235)
plt.plot(count_k.index, count_k["corr_coef"])
plt.title("휴일별 관중수와 게시글 갯수 사이 상관관계")

plt.subplot(236)
plt.plot(search_k.index, search_k["corr_coef"])
plt.title("휴일별 관중수와 검색량 사이 상관관계")
plt.show()
````

2. 야구 성적이외에 구단에서 실시하는 이벤트 또한 야구 관중들에게 주는 보상의 일종으로 야구 관객수에 영향을 줄 있는 요인중 하나이다. 이에 따라, 차후 연구에서는 야구 이벤트 특성을 고려한 예측 모형을 제안할 필요성이 있다.

3. 추후에 회귀분석보다 설명력을 올리기 위해서 딥러닝을 이용한 관중수 예측을 실시할 예정이다.

##### 기타 1 (하이퍼 파라미터)

````python
from sklearn.model_selection import KFold
scores = np.zeros(10)
cv = KFold(5, shuffle=True, random_state=22)
total_ls = []
for i, (idx_train, idx_test) in enumerate(cv.split(df)):
    df_train = df.iloc[idx_train]
    df_test = df.iloc[idx_test]
    model2 = sm.OLS.from_formula("attendance ~ C(month) + C(weekday) + C(weekend) + C(rainy) + scale(mean_c) + scale(O3) + scale(PM10) + scale(PM25) + scale(home_away) + C(win_shift) + C(month):C(weekend) +C(weekend):scale(cumsum_win) + C(month):scale(cumsum_win) + C(month):scale(m_count) + C(month):scale(search) + C(weekend):scale(m_count) + C(weekend):scale(search) + I(mean_c ** 2) + I(m_count ** 2) + I(search ** 2) + I(O3 ** 2) + I(PM10 ** 2) + I(PM25 ** 2) + I(mean_c ** 3) + I(m_count ** 3) + I(search ** 3) + I(O3 ** 3) + I(PM10 ** 3) + I(PM25 ** 3) + I(mean_c ** 4) + I(m_count ** 4) + I(search ** 4) + I(O3 ** 4) + I(PM10 ** 4) + I(PM25 ** 4) + I(mean_c ** 5) + I(m_count ** 5) + I(search ** 5) + I(O3 ** 5) + I(PM10 ** 5) + I(PM25 ** 5) + I(mean_c ** 6) + I(m_count ** 6) + I(search ** 6) + I(O3 ** 6) + I(PM10 ** 6) + I(PM25 ** 6)", data=df)
    result1 = model2.fit_regularized(alpha=0.01, L1_wt=0.5)
    pred = result1.predict(df_train)
    ls = []
    for num in range(len(pred)):
        ls.append(np.abs(1 - pred.reset_index(drop=True)
                         [num] / df_train.attendance.reset_index(drop=True)[num]))
    result_t_train = np.mean(ls)
    pred = result1.predict(df_test)
    ls = []
    for num in range(len(pred)):
        ls.append(np.abs(1 - pred.reset_index(drop=True)
                         [num] / df_test.attendance.reset_index(drop=True)[num]))
    result_t_test = np.mean(ls)
    print("학습 error = {:.4f}, 검증 error = {:.4f}".format(
        result_t_train, result_t_test))
````
